```python
import numpy as np
import matplotlib.pyplot as plt

# Part 1: Plotting the original data and best fit line
print("--- Part 1 ---")
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m:.4f}, Intercept: {c:.4f}")

plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, m*X + c, color='red', label=f'Best fit: y = {m:.2f}x + {c:.2f}')
plt.title("Part 1: Least Squares Best Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()


```

```python
# Part 2: Randomize the vectors and rerun
print("\n--- Part 2 ---")
X_rand = np.random.rand(5) * 5
Y_rand = np.random.rand(5) * 5

A_rand = np.vstack([X_rand, np.ones(len(X_rand))]).T
m_rand, c_rand = np.linalg.lstsq(A_rand, Y_rand, rcond=None)[0]
print(f"Random Slope: {m_rand:.4f}, Random Intercept: {c_rand:.4f}")

plt.figure(figsize=(8, 5))
plt.scatter(X_rand, Y_rand, color='purple', label='Random Data')
x_line = np.linspace(0, 5, 100)
plt.plot(x_line, m_rand*x_line + c_rand, color='red', label=f'Best fit: y = {m_rand:.2f}x + {c_rand:.2f}')
plt.title("Part 2: Fit on Randomized Vectors")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Part 3: Custom linear equation with random perturbation
print("\n--- Part 3 ---")
# Define original linear equation: y = 4x + 1
true_m = 4.0
true_c = 1.0

# Compute x (0 to 1, steps of .01)
X_custom = np.arange(0, 1.01, 0.01)

# Compute y with a random perturbation (noise)
noise = np.random.normal(0, 0.5, len(X_custom))
Y_custom = (true_m * X_custom) + true_c + noise

# Compute the Least Squares
A_custom = np.vstack([X_custom, np.ones(len(X_custom))]).T
m_custom, c_custom = np.linalg.lstsq(A_custom, Y_custom, rcond=None)[0]
print(f"Fitted Slope: {m_custom:.4f} (Expected ~{true_m})")
print(f"Fitted Intercept: {c_custom:.4f} (Expected ~{true_c})")

plt.figure(figsize=(8, 5))
plt.scatter(X_custom, Y_custom, color='gray', alpha=0.5, label='Noisy Data', s=15)
plt.plot(X_custom, true_m * X_custom + true_c, color='blue', linewidth=2, label='Original Equation')
plt.plot(X_custom, m_custom * X_custom + c_custom, color='red', linestyle='--', linewidth=2, label='Best Fit Line')
plt.title("Part 3: Custom Equation with Random Perturbation")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
