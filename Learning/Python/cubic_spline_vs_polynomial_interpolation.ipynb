```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Define known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# 1. Cubic Spline Interpolation
cs = CubicSpline(x, y, bc_type='natural')
x_interp = np.linspace(0, 5, 100)
y_cs_interp = cs(x_interp)

# 2. Polynomial Fitting (Degree 5 for 6 points)
poly_coeffs = np.polyfit(x, y, 5)
poly_func = np.poly1d(poly_coeffs)
y_poly_interp = poly_func(x_interp)

# Plotting both on the same graph
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'ko', markersize=8, label="Original Data Points")
plt.plot(x_interp, y_cs_interp, 'r-', linewidth=2, label="Cubic Spline (Piecewise)")
plt.plot(x_interp, y_poly_interp, 'b--', linewidth=2, label="Polynomial Fit (Degree 5)")

plt.title("Cubic Spline vs. Polynomial Fitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
