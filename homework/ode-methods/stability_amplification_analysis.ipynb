```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the domain for x
x = np.linspace(0, 3 * np.pi, 1000)

# 2. Define the time-varying lambda and the amplification factor
def lambda_x(x):
    return -2 * np.cos(x)

def amplification_factor(h, x):
    # Amplification factor = 1 + h * lambda
    return 1 + h * lambda_x(x)

# 3. Test different step sizes: a stable one, the absolute limit, and an unstable one
h_values = [0.5, 1.0, 1.2]
colors = ['green', 'blue', 'red']

# 4. Prepare the plot
plt.figure(figsize=(12, 6))

# Plot the absolute stability bounds (+1 and -1)
plt.axhline(1, color='black', linestyle='--', linewidth=1.5, label='Upper Bound (+1)')
plt.axhline(-1, color='black', linestyle='--', linewidth=1.5, label='Lower Bound (-1)')

# Shade the critical decay regions where lambda(x) < 0 (i.e., cos(x) > 0)
# In these regions, the numerical method will blow up if the amplification factor escapes [-1, 1]
plt.fill_between(x, -3, 3, where=(np.cos(x) > 0), color='gray', alpha=0.2, 
                 label='Critical Decay Region ($\lambda < 0$)')

# Plot the amplification factor curve for each h
for h, color in zip(h_values, colors):
    plt.plot(x, amplification_factor(h, x), label=f'h = {h}', color=color, linewidth=2)

# 5. Format and display
plt.ylim(-2, 3)
plt.xlim(0, 3 * np.pi)
plt.xlabel('x (radians)')
plt.ylabel('Local Amplification Factor')
plt.title("Stability Analysis: Local Amplification Factor for y' = -2y*cos(x)")
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
