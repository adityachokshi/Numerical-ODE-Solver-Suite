```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the differential equation and the exact solution
def dy_dx(x, y):
    return -2 * y * np.cos(x)

def exact_solution(x):
    return np.exp(-2 * np.sin(x))

# 2. Setup parameters
x_start = 0
x_end = 10 * np.pi
# Create the array of h values: [0.01, 0.05, 0.15, 0.20, 0.25 ... 1.0]
h_values = [0.01, 0.05] + list(np.arange(0.15, 1.05, 0.05))

# Lists to store our metrics for comparison
avg_y_vals = []
avg_errors = []
y_at_15_vals = []
local_errors_15 = []

# Prepare the 3-plot figure
plt.figure(figsize=(18, 5))

# --- PLOT 1: Solution Curves ---
plt.subplot(1, 3, 1)

# Plot the exact solution as a solid black line for baseline
x_exact = np.linspace(x_start, x_end, 1000)
plt.plot(x_exact, exact_solution(x_exact), color='black', linewidth=2, label='Exact', zorder=10)

print(f"{'h':<6} | {'Avg y':<10} | {'Avg Error':<10} | {'y at x≈15':<12} | {'Local Error at x≈15':<15}")
print("-" * 65)

# Run Euler's Method for each h
for h in h_values:
    x = np.arange(x_start, x_end + h, h)
    y = np.zeros(len(x))
    y[0] = 1.0 # Initial condition
    
    # Euler loop
    for i in range(1, len(x)):
        y[i] = y[i-1] + h * dy_dx(x[i-1], y[i-1])
        
    # Get exact values at these specific x points to calculate error
    y_exact_points = exact_solution(x)
    
    # Calculate Average y and Average Error
    avg_y = np.mean(y)
    avg_error = np.mean(np.abs(y - y_exact_points))
    
    # Find the value close to x = 15
    idx_15 = np.argmin(np.abs(x - 15))
    actual_x_15 = x[idx_15]
    y_15 = y[idx_15]
    exact_15 = y_exact_points[idx_15]
    local_error_15 = np.abs(y_15 - exact_15)
    
    # Store metrics
    avg_y_vals.append(avg_y)
    avg_errors.append(avg_error)
    y_at_15_vals.append(y_15)
    local_errors_15.append(local_error_15)
    
    # Print the comparisons
    print(f"{h:<6.2f} | {avg_y:<10.4f} | {avg_error:<10.4f} | {y_15:<12.4f} | {local_error_15:<15.4f}")
    
    # Plot a subset of h values to keep the graph readable
    if h in [0.01, 0.15, 0.5, 1.0]:
        plt.plot(x, y, alpha=0.7, label=f'h={h:.2f}')
    else:
        plt.plot(x, y, alpha=0.2, color='gray') # Plot the rest in faint gray

plt.title("Euler's Method vs Exact Solution")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.grid(True)

# --- PLOT 2: Average Error vs h ---
plt.subplot(1, 3, 2)
plt.plot(h_values, avg_errors, marker='o', color='red', linestyle='-')
plt.title("Average Error vs Step Size (h)")
plt.xlabel("Step Size (h)")
plt.ylabel("Average Absolute Error")
plt.grid(True)

# --- PLOT 3: Local Error at x ≈ 15 vs h ---
plt.subplot(1, 3, 3)
plt.plot(h_values, local_errors_15, marker='s', color='blue', linestyle='-')
plt.title("Local Error at x ≈ 15 vs Step Size (h)")
plt.xlabel("Step Size (h)")
plt.ylabel("Local Error")
plt.grid(True)

plt.tight_layout()
plt.show()
```
