```python
import numpy as np
import matplotlib.pyplot as plt

def solve_homework():
    # Define Matrix A and Vector b 
    A = np.array([[3, -0.1, -0.2],
                  [0.1, 7, -0.3],
                  [0.3, -0.2, 10]])
    b = np.array([7.85, -19.3, 71.4])
    
    # Initial parameters [cite: 1130, 1132]
    x0 = np.array([0, 0, 0])
    tolerances = [1e-3, 1e-6, 1e-9]
    
    results = {}

    for method in ['Jacobi', 'Gauss-Seidel']:
        results[method] = {}
        for tol in tolerances:
            x = x0.astype(float).copy()
            errors = []
            n = len(b)
            
            for k in range(1, 101): # Max 100 iterations
                x_old = x.copy()
                if method == 'Jacobi':
                    # Jacobi updates using values from previous iteration [cite: 789, 792]
                    x_new = np.zeros(n)
                    for i in range(n):
                        s = sum(A[i, j] * x_old[j] for j in range(n) if i != j)
                        x_new[i] = (b[i] - s) / A[i, i]
                    x = x_new
                else:
                    # Gauss-Seidel updates in-place immediately [cite: 903, 1024]
                    for i in range(n):
                        s1 = sum(A[i, j] * x[j] for j in range(i))
                        s2 = sum(A[i, j] * x_old[j] for j in range(i + 1, n))
                        x[i] = (b[i] - s1 - s2) / A[i, i]
                
                # Calculate absolute error between iterations [cite: 729, 1075]
                error = np.linalg.norm(x - x_old, ord=np.inf)
                errors.append(error)
                
                if error < tol:
                    results[method][tol] = (k, errors)
                    break
                    
    return results

# Execute and prepare for plotting
homework_results = solve_homework()
```
